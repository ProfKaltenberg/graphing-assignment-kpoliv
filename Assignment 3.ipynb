```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("final_df.csv")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14,6))

# Graph 1: Votes
top10_byvotes = df.sort_values(by="votes", ascending=True).tail(10)
top10_byvotes
ax1.barh(top10_byvotes['title'], top10_byvotes['votes'], color='#2E5B88')
ax1.set_title("Top 10 Books by Votes")
ax1.set_xlabel("Number of Votes")

# Graph 2: Recommendations
top10_rec = df.sort_values(by="recommendations", ascending=True).tail(10)
ax2.barh(top10_rec['title'], top10_rec['recommendations'], color='#2E5B88')
ax2.set_title("Top 10 Books by Recommendations")
ax2.set_xlabel("Number of Recommendations")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(data=df, x="ranking", y="votes", alpha=0.5, color='#2E5B88')
plt.title("Votes vs Ranking")
plt.show()
```

## Graphs interpretations. 

The bar graphs suggest that there is no consistent relationship between a book’s position in the dataset and the number of votes or recommendations it receives. When comparing the two bar charts, we can see that the ranking of books changes across them, meaning that books that appear at the top in one graph do not necessarily remain at the top in the other. 

Regarding to the scatter plot there is no clear linear relationship between the variables. While top-ranked books tend to include some of the most voted titles, there is significant variation within this group. Also, lower-ranked books show more consistent vote counts, indicating that other factors diffetnet than votes may play an important role in determining ranking.
